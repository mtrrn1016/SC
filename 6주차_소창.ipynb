## 중첩 if문을 활용한 문제 만들어줘

### 문제: 놀이공원 입장료 계산하기

나이와 키를 입력받아 놀이공원 입장료를 계산하는 프로그램을 작성하세요.

- 나이가 8세 미만이거나 65세 이상이면 무료입니다.
- 나이가 8세 이상이고 65세 미만인 경우, 키가 120cm 이상이면 10000원, 120cm 미만이면 5000원입니다.

```python
age = int(input("나이를 입력하세요: "))
height = int(input("키를 입력하세요 (cm): "))

if age < 8 or age >= 65:
  print("입장료는 무료입니다.")
else:
  if height >= 120:
    print("입장료는 10000원입니다.")
  else:
    print("입장료는 5000원입니다.")
```

## 맥주 구매에 대한 연령제한을 가지고 if문을 활용해 문제내줘

### 문제: 맥주 구매 가능 여부 판별하기

나이를 입력받아 맥주 구매가 가능한지 여부를 판별하는 프로그램을 다중 if문을 사용하여 작성하세요.

- 19세 이상이면 "맥주 구매 가능"을 출력합니다.
- 19세 미만이면 "미성년자는 맥주를 구매할 수 없습니다."를 출력합니다.

```python
age = int(input("나이를 입력하세요: "))

if age >= 19:
  print("맥주 구매 가능")
else:
  print("미성년자는 맥주를 구매할 수 없습니다.")
```

## 다중if문을 활용해 이전보다 더 어렵게 문제를 내줘

### 문제: 학점 계산하기

시험 점수를 입력받아 다음과 같이 학점을 출력하는 프로그램을 다중 if문을 사용하여 작성하세요.

- 95점 이상: A+
- 90점 이상 95점 미만: A
- 85점 이상 90점 미만: B+
- 80점 이상 85점 미만: B
- 75점 이상 80점 미만: C+
- 70점 이상 75점 미만: C
- 65점 이상 70점 미만: D+
- 60점 이상 65점 미만: D
- 60점 미만: F

```python
score = int(input("시험 점수를 입력하세요: "))

if score >= 95:
  print("A+")
elif score >= 90:
  print("A")
elif score >= 85:
  print("B+")
elif score >= 80:
  print("B")
elif score >= 75:
  print("C+")
elif score >= 70:
  print("C")
elif score >= 65:
  print("D+")
elif score >= 60:
  print("D")
else:
  print("F")
```

## 다중 if문인데 주어진 조건에 맞게 코드 작성하는 유형의 문제로 내줘


### 문제: 대중교통 요금 계산하기

나이와 탑승 시간에 따라 대중교통 요금을 계산하는 프로그램을 다중 if문을 사용하여 작성하세요.

- 0세 이상 7세 이하: 무료
- 8세 이상 19세 이하: 800원
- 20세 이상 64세 이하: 1200원
- 65세 이상: 무료

단, 탑승 시간이 밤 10시 이후부터 새벽 4시 이전까지인 경우 (22시 ~ 3시 59분), 20세 이상 64세 이하 승객은 1000원의 할인 요금을 적용받습니다.

```python
age = int(input("나이를 입력하세요: "))
time = int(input("현재 시간을 입력하세요 (0~23): "))

if age <= 7 or age >= 65:
  print("요금은 무료입니다.")
elif 8 <= age <= 19:
  print("요금은 800원입니다.")
elif 20 <= age <= 64:
  if time >= 22 or time < 4:
    print("요금은 1000원입니다.")
  else:
    print("요금은 1200원입니다.")
else:
  print("잘못된 나이 입력입니다.")
```
