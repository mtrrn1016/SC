# 1. 사용자로부터 10개의 단어를 입력 받아 역순으로 출력

```python
print("**** 10개의 단어를 입력받아 역순으로 출력 ****")
words = []
for i in range(10):
    word = input(f"단어입력({i+1}) : ")
    words.append(word)

for word in reversed(words):
    print(word)
```

# 2. 사용자로부터 10개의 정수를 입력 받아 역순으로 양수만 출력

```python
print("**** 10개의 정수를 입력받아 역순으로 양수만 출력 ****")
numbers = []
for i in range(10):
   num = int(input(f"{i+1}. 정수입력 : "))
   numbers.append(num)

for num in reversed(numbers):
    if num > 0:
        print(num)
```

# 3. 사용자로부터 n개의 정수를 입력 받아 역순으로 양수만 출력

```python
print("**** n개의 정수를 입력 받아 역순으로 양수만 출력 ****")

while True:
  n = int(input("몇개의 정수를 입력합니까? : "))
  if n < 1:
    print("n은 1 이상의 정수여야 합니다. 다시 입력해주세요.")
    continue
  break

numbers = []
for i in range(n):
  num = int(input(f"({i+1}/{n}) 정수입력 : "))
  numbers.append(num)

for num in reversed(numbers):
    if num > 0:
        print(num)
```

# 4. 사용자로부터 n개의 정수를 입력받아 인덱스가 홀수 위치의 요소 중 짝수만 출력

```python
print("**** n개의 정수를 입력 받아 인덱스가 홀수 위치의 요소 중 짝수만 출력 ****")

while True:
  n = int(input("몇개의 정수를 입력합니까? : "))
  if n < 1:
    print("n은 1 이상의 정수여야 합니다. 다시 입력해주세요.")
    continue
  break

numbers = []
for i in range(n):
  num = int(input(f"정수 입력({i+1}/{n}) : "))
  numbers.append(num)

for i in range(len(numbers)):
    if (i % 2 != 0) and (numbers[i] % 2 == 0):
        print(numbers[i])
```

# 5. 사용자로부터 n개의 점수를 입력 받아 합과 평균을 출력하기

```python
print("**** n개의 점수를 입력 받아 합과 평균을 출력하기 ****")

while True:
  n = int(input("몇개의 과목의 점수를 입력합니까? : "))
  if n < 1:
    print("잘못된 수 입니다. 다시 입력해주세요.")
    continue
  break

scores = []
for i in range(n):
  score = int(input(f"{i+1} 과목 점수 : "))
  scores.append(score)

total_sum = sum(scores)
average = total_sum // n if n > 0 else 0

print(f"합 : {total_sum}, 평균 : {average}")
```

# 6. 헌혈현황관리 프로그램[리스트] - 10개 입력

```python
blood_types = {}
valid_blood_types = ['A', 'B', 'AB', 'O']

for i in range(10):
    while True:
        blood_type = input(f"헌혈해 주셔서 감사합니다. 헌혈하신 혈액형을 선택하세요 \nA, B, AB, O : ").upper()
        if blood_type in valid_blood_types:
            break
        else:
            print("잘못된 혈액형입니다. (A, B, AB, O 중 하나를 입력해주세요)")

    if blood_type in blood_types:
        blood_types[blood_type] += 1
    else:
        blood_types[blood_type] = 1
print("----------------------------------")
print("혈액형 : 개수 ")
print("----------------------------------")
for bt, count in blood_types.items():
    print(f"{bt}형 : {count}")
print("----------------------------------")
```

# 7. 숫자맞추기 게임[변수, random()]

```python
import random
target_number = random.randint(1, 100)

while True:
  guess = int(input("1부터 100 사이의 숫자를 맞춰보세요: "))
  if guess < target_number:
     print("너무 작아요")
  elif guess > target_number:
      print("너무 커요")
  else:
      print(f"정답!")
      break
```

# 8. 끝말잇기 게임[변수, len()]

```python
words_history = set()
current_word = "가나다"

print("끝말잇기 게임")
print("시작")
print(f"{current_word}")
words_history.add(current_word)

while True:
  user_word = input(f"단어 : ").strip()

  if user_word[0] != current_word[-1]:
    print(f"땡!")
    break
  if user_word in words_history:
    print(f"땡!")
    break

  words_history.add(user_word)
  current_word = user_word
```

# 9. 합산프로그램[변수]

```python
print("i부터 n까지의 합을 계산하면 얼마일까요?")

while True:
  i = int(input("연산할 시작 수를 입력하세요 : "))
  if i < 1:
    print("n은 1 이상의 자연수여야 합니다. 다시 입력해주세요.")
    continue
  break

while True:
  n = int(input("연산할 마지막 수를 입력하세요 : "))
  if n < 1:
    print("n은 1 이상의 자연수여야 합니다. 다시 입력해주세요.")
    continue
  elif n < i:
    print(f"n은 i({i})보다 크거나 같아야 합니다. 다시 입력해주세요.")
    continue
  break

total_sum = 0
for num in range(i, n + 1):
    total_sum += num

print(f"합은 {total_sum} 입니다.")
```

# 10. 카페 주문 프로그램

```python
menu = {
    "아메리카노": 2000,
    "카페라떼": 3000,
    "마끼아또": 3500
}

menu_items_list = list(menu.items())

order_list = {}
total_price = 0

while True:
    print("\n===== 슈먼카페 =====")
    for i, (item_name, price) in enumerate(menu_items_list):
        print(f"{i+1}. {item_name} : {price}원")
    print(f"{len(menu_items_list) + 1}. 주문종료\n")


    choice_input = input("메뉴를 입력해주세요 => ").strip()

    if not choice_input.isdigit():
       print("잘못 입력됐습니다. 다시 입력해주세요.")
       continue

    choice_num = int(choice_input)

    if choice_num == len(menu_items_list) + 1:
       print("주문을 종료합니다.")
       break

    elif 1 <= choice_num <= len(menu_items_list):
       selected_item_name, item_price = menu_items_list[choice_num - 1]

       while True:
           try:
               quantity = int(input("수량을 입력해주세요 => "))
               if quantity <= 0:
                   print("수량은 1 이상이어야 합니다.")
                   continue
               break
           except ValueError:
               print("잘못 입력됐습니다. 다시 입력해주세요.")

       order_list[selected_item_name] = order_list.get(selected_item_name, 0) + quantity
       total_price += item_price * quantity

    else:
      print("잘못 입력됐습니다. 다시 입력해주세요.")

    print("\n**** 총 주문내역 ****")
    if not order_list:
      print("주문하신 상품이 없습니다.")
    else:
      for i, (item_name, _) in enumerate(menu_items_list):
        quantity = order_list.get(item_name, 0)
        print(f"{i+1}. {item_name}: {quantity}잔")
      print(f"총 금액 : {total_price}원")
print("\n**** 총 주문내역 ****")
if not order_list:
  print("주문하신 상품이 없습니다.")
else:
  for i, (item_name, _) in enumerate(menu_items_list):
    quantity = order_list.get(item_name, 0)
    print(f"{i+1}. {item_name}: {quantity}잔")
  print(f"총 금액 : {total_price}원")
```
