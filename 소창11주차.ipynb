## 데이터 프레임 생성법

### 데이터 프레임에 리스트를 이용해서 1부터 9까지 3행3열로 출력하는 코드를 작성해줘

```python
import pandas as pd
# 1부터 9까지의 숫자를 담은 리스트를 사용하여 3x3 데이터프레임 생성
data = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
df = pd.DataFrame(data)

#데이터셋 출력
print(df)
display(df)
```

### 제미나이에게 내가 물어본 결과

```python
import pandas as pd
import numpy as np

# 1부터 9까지의 숫자로 이루어진 리스트 생성
data = list(range(1, 10))

# 3x3 형태로 리스트를 재구성
# pandas DataFrame은 2차원 리스트나 numpy 배열을 직접 받아들일 수 있습니다.
# 여기서는 reshape 기능을 보여주기 위해 numpy를 사용합니다.
reshaped_data = np.array(data).reshape(3, 3)

# 데이터프레임 생성
df = pd.DataFrame(reshaped_data)

# 데이터프레임 출력
display(df)
```

```python
import matplotlib.pyplot as plt # Matplotlib 라이브러리를 plt라는 별칭으로 불러옵니다.

x = [1, 2, 3, 4, 5] # x축 데이터를 정의합니다.
y = [10, 15, 13, 18, 20] # y축 데이터를 정의합니다.

plt.plot(x, y) # x와 y 데이터를 사용하여 선 그래프를 그립니다.

plt.xlabel('X-axis Label') # x축에 라벨을 추가합니다.
plt.ylabel('Y-axis Label') # y축에 라벨을 추가합니다.
plt.title('Line Plot of X vs Y') # 그래프의 제목을 설정합니다.

plt.grid(True) # 그래프에 격자를 표시합니다.
plt.show() # 그래프를 화면에 표시합니다.
```

```python
import yfinance as yf
import pandas as pd

# Tesla (TSLA) 종목 설정
ticker_symbol = 'TSLA'

# 2025년 5월 시작일과 종료일 설정
start_date = '2025-05-01'
end_date = '2025-05-31'

# Yahoo Finance에서 주가 데이터 다운로드
data = yf.download(ticker_symbol, start=start_date, end=end_date)

# 데이터 출력
print(f"\n{ticker_symbol} 주가 데이터 ({start_date} ~ {end_date}):")
display(data)
```

```python
# 'Close' 컬럼(종가) 데이터만 출력
display(data['Close'])
```

```python
import matplotlib.pyplot as plt # Matplotlib 라이브러리를 plt라는 별칭으로 불러옵니다.

# 'data' 데이터프레임에서 'Close' 컬럼 데이터를 가져옵니다.
close_prices = data['Close']

plt.figure(figsize=(10, 6)) # 그래프의 크기를 설정합니다 (가로 10인치, 세로 6인치).
plt.plot(close_prices.index, close_prices.values) # 종가 데이터를 사용하여 선 그래프를 그립니다. 인덱스를 x축, 값을 y축으로 사용합니다.

plt.xlabel('Date') # x축 라벨을 'Date'로 설정합니다.
plt.ylabel('Closing Price') # y축 라벨을 'Closing Price'로 설정합니다.
plt.title('Tesla Closing Price Trend (May 2025)') # 그래프의 제목을 'Tesla Closing Price Trend (May 2025)'로 설정합니다.

plt.grid(True) # 그래프에 격자를 표시하여 가독성을 높입니다.
plt.xticks(rotation=45) # x축 날짜 라벨이 겹치지 않도록 45도 회전합니다.
plt.tight_layout() # 그래프 요소들이 깔끔하게 배치되도록 조정합니다.
plt.show() # 그래프를 화면에 표시합니다.
```
