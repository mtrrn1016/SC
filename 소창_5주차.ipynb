## 논리 연산자
and, or, not

```python
num1= 10
num2 = 20
num3 = 15

print(f"({num1} < {num2}) and ({num2} > {num3}) : ", (num1 < num2) and (num2 > num3))
print(f"({num1} == {num2}) and ({num3} > {num1}) :", (num1 ==num2) and (num3 > num1))
print(f"({num1} < {num3}) and ({num2} <= {num3}) :", (num1 < num3) and (num2 <= num3))

print(f"({num1} < {num2}) or ({num2} > {num3}) : ", (num1 < num2) or (num2 > num3))
print(f"({num1} == {num2}) or ({num3} > {num1}) :", (num1 ==num2) or (num3 > num1))
print(f"({num1} < {num3}) or ({num2} <= {num3}) :", (num1 < num3) or (num2 <= num3))

print(f"not({num1} == {num2}) : ", not(num1 == num2))
print(f"not({num1} < {num3}) : ", not(num1 < num3))
```

## 문제: 나이와 이름으로 메시지 만들기 (빈칸 채우기)

나의 질문 : 이번엔 아까 문제에 산술 연산자와 문자열 연산자를 섞은 문제를 내줘. 문제의 유형은 빈칸넣기로 해줘

사용자로부터 나이와 이름을 입력받아 다음 조건을 확인하고, 조건에 따라 메시지를 출력하는 프로그램을 완성하세요.

**조건:**
나이가 10세 이상이고 20세 이하인가?

**출력:**
조건을 만족하면 "안녕하세요, [이름]! 당신은 [나이]살이군요. 멋진 나이입니다!" 를 출력하세요.
조건을 만족하지 않으면 "[이름]님, 죄송하지만 이 프로그램은 10세 이상 20세 이하만 사용 가능합니다." 를 [나이] 번 반복하여 출력하세요.

아래 코드의 빈칸을 채워 프로그램을 완성하세요.

```python
# 여기에 코드를 작성하세요.
age = int(input("나이를 입력하세요: "))
name = input("이름을 입력하세요: ")

if (age >= 10) and (age <= 20):
    print(f"안녕하세요, {name}! 당신은 {age}살이군요. 멋진 나이입니다!")
else:
    print(f"{name}님, 죄송하지만 이 프로그램은 10세 이상 20세 이하만 사용 가능합니다." * age)
```

## 문제: 비밀번호 확인

나의 질문 : 문자열 연산자와 논리연산자를 이용한 간단한 문제 내줘

사용자로부터 비밀번호를 입력받아 다음 두 조건을 모두 만족하는지 확인하는 프로그램을 작성하세요.

1. 비밀번호가 8자 이상인가?
2. 비밀번호에 '!' 또는 '#' 문자가 포함되어 있는가?

위 두 가지 조건을 모두 만족하면 "비밀번호가 안전합니다."를 출력하고, 하나라도 만족하지 않으면 "비밀번호가 안전하지 않습니다."를 출력하세요.

```python
# 여기에 코드를 작성하세요.
password = input("비밀번호를 입력하세요: ")

if len(password) >= 8 and ('!' in password or '#' in password):
    print("비밀번호가 안전합니다.")
else:
    print("비밀번호가 안전하지 않습니다.")
```
